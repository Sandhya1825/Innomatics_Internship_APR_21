**Univariate** **Plots:**

Univariate data visualization plots help us comprehend the enumerative properties as well as a descriptive summary of the particular data variable. These plots help in understanding the location/position of observations in the data variable, its distribution, and dispersion. Uni-variate plots are of two types: 
1. Enumerative plots and
2. Summary plots

```python
# list of numerical variables
num_feat = [feature for feature in data.columns if data[feature].dtypes != 'O']
print('Number of numerical variables: ', len(num_feat))
data[num_feat].head(3)
```

```python
plt.figure(figsize=(20,22))
n = 0
plt.rc('ytick', labelsize=10)
plt.rc('xtick', labelsize=10)
for i, column in enumerate(data[num_feat].columns):
    n+=1
    plt.subplot(3, 3, n)
    sns.distplot(data[num_feat][column], bins=30)
    plt.title('Distribution var {}'.format(column))
plt.show()
```

**Handling** **Outliers**

From the statistical study, it is found that may be present of outliers. This can be confirmed by the following graphs.

```python
for column in data:
    
    if data[column].dtypes != 'O':
        #before
        plt.figure(figsize=(10,8))
        plt.subplot(1,2,1)
        sns.boxplot(data[column])
        plt.title("Boxplot before median impuation")
        plt.show()
        
        quar_1 = data[column].quantile(0.25)
        quar_3 = data[column].quantile(0.75)
        
        iqr = quar_3 - quar_1
        
        l_tail = quar_1 - 1.5 * iqr
        u_tail = quar_3 + 1.5 * iqr
        
        median_imp = np.median(data[column])
        
        for i in data[column]:
            if i> u_tail or i < l_tail:
                data[column] = data[column].replace(i,median_imp)
                
        #after
        plt.figure(figsize=(10,8))
        plt.subplot(1,2,2)
        sns.boxplot(data[column])
        plt.title("Boxplot after median imputation")
        plt.show()
        
    else:
        pass
```
